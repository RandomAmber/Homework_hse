```python
#Определить грамматические показатели для прилагательных в польском языке. Тренировка на циклы и условия. 
#Цикл бесконечный, пока пользователь не введет пустое слово. 

#s = ''.join([i for i in s if not i.isdigit()]) - если ввели числа
#word = ''.join(e for e in s if e.isalnum()) - если ввели специальные символы
#s = s.strip() - удалит лишние пробелы
#s = s.lower() - все к нижнему регистру


#Код программы:

def superprint():
    return print ("It's definitely an adjective. Must be one of these:")


while True:
    s = str(input())
    s = s.lower()
    s = s.strip()
    s = ''.join([i for i in s if not i.isdigit()])
    word = ''.join(e for e in s if e.isalnum())
    if word == "":
        break
    if word[-1] == "y":
        superprint()
        print ("Nom.Sg.Masc.Inan. / Voc.Sg.Masc.Inan. / Nom.Sg.Masc.Anim. / Acc.Sg.Masc.Inan. / Voc.Sg.Masc.Anim.")
    elif word.endswith("ego"):
        superprint()
        print ("Gen.Sg.Masc.Inan. / Gen.Sg.Masc.Anim.  / Gen.Sg.Neut.")
    elif word.endswith('ym'):
        superprint()
        print ("Inst.Sg.Masc.Inan. / Loc.Sg.Masc.Inan.  / Inst.Sg.Masc.Anim. / Loc.Sg.Masc.Anim. / Inst.Sg.Neut. / Loc.Sg.Neut.")
    elif word.endswith('emu'):
        superprint()
        print ("Dat.Sg.Masc.Inan. / Dat.Sg.Masc.Anim. / Dat.Sg.Neut.")
    elif word[-1] == "e":
        superprint()
        print ("Nom.Sg.Neut. / Voc.Sg.Neut. / Acc.Sg.Neut. / Nom.Pl.Impers. / Acc.Pl.Impers. / Voc.Pl.Impers.")
    elif word.endswith('a'):
        superprint()
        print ("Nom.Sg.Fem. / Voc.Sg.Fem")
    elif word.endswith('ą'):
        superprint()
        print ("Acc.Sg.Fem. / Inst.Sg.Fem.")
    elif word.endswith('ej'):
        superprint()
        print ("Loc.Sg.Fem. / Gen.Sg.Fem. / Dat.Sg.Fem.")
    elif word.endswith('i'):
        superprint()
        print ('Nom.Pl.Masc.Pers. / Voc.Pl.Masc.Pers.')
    elif word.endswith('ych'):
        superprint()
        print ('Acc.Pl.Masc.Pers. / Loc.Pl.Masc.Pers. / Gen.Pl.Masc.Pers. / Loc.Pl.Impers. / Gen.Pl.Impers.')
    elif word.endswith ('ymi'):
        superprint()
        print ('Inst.Pl.Masc.Pers. / Inst.Pl.Impers.')
    elif word.endswith ('ym'):
        superprint()
        print ("Dat.Pl.Masc.Pers. / Dat.Pl.Impers.")
    else:
        print ("No idea, sorry")
```
