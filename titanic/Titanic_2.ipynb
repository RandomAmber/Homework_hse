```python
train_data.info()
```

Есть много способов обрабатывать пропуски. 

Например: Интерполяция, ближайший сосед, новое значение, самое частое значение и т.д. Советую этот момент посмотреть в статьях на Хабре.

К слову, советую на каждую функцию заглядывать в документацию и смотреть какие у нее есть параметры. Параметры функции - то, что мы указываем в скобках. Могут быть обязательными и нет. Если в документации, например, указано что-то такое:

func(number, count=10, save_path=None)

Здесь number - то, что тебе точно нужно подать на вход функции, и без этого никак. Count - ты можешь указать, но у него есть значение по умолчанию - 10. Save_path - также можно указать, а можно и нет, но у него нет значения по умолчанию. Значит параметр вцелом опциональный. Что означает каждый параметр обычно написано чуть ниже.

Вот, например, у интерполяции есть даже разные варианты способов интерполяции - http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html

### Encoding

Если перезапустить ячейку снизу, она не заработает. Нужно установить sklearn и добавить подключение библиотеки наверху (можно и здесь, но удобно выносить все библиотеки наверх). Обрати внимание, что обращаясь к LabelEncoder я не пишут sklearn.preprocessing.LabelEncoder

```python
#создаем объект типа LabelEncoder
le = LabelEncoder() 
#обучаем его, он должен видеть какие бывают значения
le.fit(train_data.Sex)
#теперь используем, чтобы трансформировать наши данные
train_data.Sex = le.transform(train_data.Sex)
```

# Подбор параметров (тюнинг)

Основные варианты - перебор по сетке или рандомезированный перебор по сетке. В обоих случаях мы указываем параметры которые хотим попробовыать и способ оценки модели. Но во втором случае проверяются не все параметры, а только какая-то часть. Причем алгоритм ищет наиболее подходящий вариант следующим способом: например, если увеличивать количество деревьев в лесу вдруг лучше уже не становится, тогда алгоритм перестанет шагать по этому параметру, даже если у нас еще остались варианты.
