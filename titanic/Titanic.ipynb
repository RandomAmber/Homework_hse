## Managing environments and install libraries

Для того, чтобы работать над разными проектами или не засорить ноут десятью тысячами ненужных библиотек советую настроить на ноутбуке окружение. Если ты работаешь с anaconda (это менеджер пакетов), то это с этим можно разобраться по этой ссылке:

https://conda.io/docs/user-guide/tasks/manage-environments.html#creating-an-environment-with-commands

Ты создаешь окружение командой
conda create --name myenv

Myenv можно заменить на любое название. Например,
conda create --name titanic

После этого сразу активируешь его командой: 
source activate titanic

Как только набираешь ее, ты находишься внутри этого окружения (именно в этом окне терминала, в других - нет!).

Теперь можно устанавливать библиотеки. Тебе точно понадобится pandas и numpy. Набери команды..

pip install pandas

pip install numpy

Теперь заходи в ноутбук. Но именно из окна терминала, где включено окружение. Подробнее про то, как его отключать и так далее читай в ссылке выше.

# Data preprocessing

### Encoding

Если перезапустить ячейку снизу, она не заработает. Нужно установить sklearn и добавить подключение библиотеки наверху (можно и здесь, но удобно выносить все библиотеки наверх). Обрати внимание, что обращаясь к LabelEncoder, я не пишу sklearn.preprocessing.LabelEncoder

```python
#создаем объект типа LabelEncoder
le = LabelEncoder() 
#обучаем его, он должен видеть какие бывают значения
le.fit(train_data.Sex)
#теперь используем, чтобы трансформировать наши данные
train_data.Sex = le.transform(train_data.Sex)
```

Проверим, сработало ли.

```python
train_data.head(2)
```

Да, sex теперь число. Можно посмотреть какие значения видел LabelEncoder.

```python
print(le.classes_)
```

кстати, лучше не совать в трансформер значения, которые LabelEncoder не видел в обучении. Тогда он не знает как их кодировать и выдает ошибку.

```python
train_data.Embarked.value_counts()
```

```python
train_data.Embarked = train_data.Embarked.astype(str)
```
